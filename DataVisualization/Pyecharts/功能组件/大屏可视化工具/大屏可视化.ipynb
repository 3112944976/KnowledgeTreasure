```python
import pandas as pd
import numpy as np
from pyecharts import options as opts 
from pyecharts.charts import Map,Bar,Line,Page,Pie,Timeline,Gauge,Map3D,Tab,Grid
from pyecharts.globals import ChartType, SymbolType,ThemeType
from pyecharts.commons.utils import JsCode
```

```python
#图一
def aa():
    tmp = r'C:\Users\1700\Desktop\信息可视化\源数据\国家统计局年度数据\国民经济核算\国内生产总值.csv'
    obj1 = pd.read_csv(tmp)
    x = np.array(obj1['年份'])
    y1 = np.array(obj1['国民总收入'])
    y2 = np.array(obj1['第一产业增加值'])
    y3 = np.array(obj1['第二产业增加值'])
    y4 = np.array(obj1['第三产业增加值'])
    y5 = np.array(obj1['人均国内生产总值'])
    x = [str(i)+'年' for i in x[::-1]]
    def set_y(x):
        tmp = [int(i/1000) for i in x[::-1]]
        return tmp
    a = (
        Bar(
            init_opts=opts.InitOpts(
                theme=ThemeType.VINTAGE,
                #bg_color={"type": "pattern", "image": JsCode("img"), "repeat": "no-repeat"}#背景图片配置
            )
        )
        .add_xaxis(x)
        .add_yaxis("国民总收入", set_y(y1), stack="stack1")
        .add_yaxis("第一产业增加值",set_y(y2), stack="stack1")
        .add_yaxis("第二产业增加值", set_y(y3), stack="stack1")
        .add_yaxis("第三产业增加值", set_y(y4), stack="stack1")
        .add_yaxis("人均国内生产总值", set_y(y5), stack="stack1")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="China"),
                        datazoom_opts=opts.DataZoomOpts(),
    #                     toolbox_opts=opts.ToolboxOpts(),
    #                     legend_opts=opts.LegendOpts(is_show=False)
                        )
        #.render("融合式条形图.html")
    )
#     a.add_js_funcs(
#         """
#         var img = new Image(); img.src = '123.jpg';
#         """
#     )#背景图片配置
    return a
```

```python
#图二
def bb():
    tmp = r'C:\Users\1700\Desktop\信息可视化\源数据\国家统计局分省年度数据\国民经济核算\地区生产总值指数.csv'
    obj1 = pd.read_csv(tmp)
    obj2 = obj1[obj1['年份'].isin([2019])]
    x = [i[:2] for i in obj2['省份']]
    x[x.index('内蒙')] = '内蒙古'
    x[x.index('黑龙')] = '黑龙江'

    y = [round((i-87)*10) for i in obj1['地区生产总值指数(上年=100)']]
    y_2 = []
    for i in range(0,31):
        y_2.append(list(y[0:20]))
        y = y[20:]
    y_3 = np.array(y_2).T.tolist()

    tl = Timeline()
    count = 0
    for i in range(2019,1999,-1):
        map0 = (
            Map(init_opts=opts.InitOpts(bg_color="dimgray", theme=ThemeType.DARK))
            .add("商家A", [list(z) for z in zip(x, y_3[count])], "china")
            .set_global_opts(
                title_opts=opts.TitleOpts(title="中国{}年地区生产总值指数热点地图".format(i)),
                visualmap_opts=opts.VisualMapOpts(min_=160, 
                                                          max_=280, 
                                                          is_piecewise=True, )
            )
        )
        count += 1
        tl.add(map0, "{}年".format(i))
        tl.add_schema(is_auto_play=True,is_rewind_play=True)
    return tl
```

```python
#图三
def cc():
    tmp = r'C:\Users\1700\Desktop\信息可视化\源数据\国家统计局年度数据\财政\中央和地方财政收入及比重.csv'
    obj1 = pd.read_csv(tmp)

    x = [str(round(i))+'年' for i in list(np.array(obj1['年份']))]
    y1 = list(np.array(obj1['全国财政收入']))
    y2 = list(np.array(obj1['中央财政收入']))
    y3 = list(np.array(obj1['地方财政收入']))

    def round_set(x):
        x = [round(i, 0) for i in x]
        return x
    round_set(y1)
    round_set(y2)
    round_set(y3)

    bar = (
        Bar(init_opts=opts.InitOpts(width="1000px", 
                                    height="500px", 
                                    #bg_color="dimgray", 
                                    theme=ThemeType.VINTAGE,
                                    animation_opts=opts.AnimationOpts(animation_delay=1000, animation_easing="elasticOut")
                                   )
           )
        .add_xaxis(xaxis_data=x[::-1])
        .add_yaxis(
            series_name="中央财政收入",
            y_axis=y2[::-1],
            yaxis_index=0,
            #color=colors[1]
        )
        .add_yaxis(series_name="地方财政收入", y_axis=y3[::-1], yaxis_index=1)#, color=colors[0]
        .extend_axis(
            yaxis=opts.AxisOpts(
                name="中央财政收入",
                type_="value",
                min_=6000,
                max_=200000,
                position="right",
                axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#F75000")),
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            )
        )
        .extend_axis(
            yaxis=opts.AxisOpts(
                type_="value",
                name="地方财政收入",
                min_=6000,
                max_=200000,
                position="left",
                axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#F75000")),
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
                splitline_opts=opts.SplitLineOpts(is_show=True, linestyle_opts=opts.LineStyleOpts(opacity=1)),
            )
        )
        .set_series_opts(itemstyle_opts={"normal": {"barBorderRadius": [30, 30, 30, 30],}})
        .set_global_opts(
            title_opts=opts.TitleOpts(title="中国财政收入可视化"),#, subtitle="我是副标题"
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],


            toolbox_opts=opts.ToolboxOpts(is_show=True),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),

            yaxis_opts=opts.AxisOpts(
                type_="value",
                name="",
                min_=6000,
                max_=200000,
                position="right",
                #offset=80,
                axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#0066CC")),
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            ),
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
        )
    )
    line = (
        Line()
        .add_xaxis(xaxis_data=x[::-1])
        .add_yaxis(series_name="全国财政收入", y_axis=y1[::-1], yaxis_index=2)#, color="#CE0000"
    )
    #bar.overlap(line).render("财政收入可视化.html")
    #bar.overlap(line)
    return bar.overlap(line)
```

```python
#图四
def dd():
    url = r"C:\Users\1700\Desktop\信息可视化\源数据\国家统计局年度数据\国民经济核算\分行业增加值.csv"
    obj1 = pd.read_csv(url)
    obj2 = obj1.set_index('年份')
    obj2.columns=['年份','农林牧渔业','工业','建筑业','批发和零售业','交通运输、仓储和邮政业','住宿和餐饮业','金融业','房地产业','其他行业']
    tl = Timeline()
    count = 0
    for i in range(2019, 1999, -1):
        tmp = []
        x = obj2.columns.tolist()[1:]
        y = [y for y in dict(obj2.iloc[count,]).values()]
        for a,b in zip(x,y):
            tmp.append((a,b))
        pie = (
            Pie(init_opts=opts.InitOpts(width="1600px", height="800px", bg_color="#2c343c"))
            .add(
                "",
                tmp,
                rosetype="radius",
                radius=["30%", "55%"],
            )
            .set_global_opts(title_opts=opts.TitleOpts(title="{0}年分行业增加值动态饼状图".format(i),
                                                       pos_left="center",
                                                       pos_top="50",
                                                       title_textstyle_opts=opts.TextStyleOpts(color="#FF8040"),
                                                      ),
                            )
        )
        count += 1
        tl.add(pie, "{}年".format(i))
        tl.add_schema(is_auto_play=True,is_rewind_play=True,play_interval=500)
    #tl.render("timeline_pie.html")
    return tl
```

```python
#图五
def ee():
    tl = Timeline()
    for i in range(2000,2020):
        g=(
            Gauge(init_opts=opts.InitOpts(width="1000px",height="600px"))
            .add(""
                 ,[('',i)]
                 ,radius="100%"
                 ,min_=2000
                 ,max_=2019
                 ,split_number=19
                )
            .set_global_opts(title_opts=opts.TitleOpts(title="时间刻度量规：{0}年".format(i)))
        )
        tl.add(g, "{}年".format(i))
        tl.add_schema(is_auto_play=True,is_timeline_show=False,play_interval=500)
    #tl.render("timeline_pie.html")
    return tl
```

```python
#图六
def ff():
    tmp = r'C:\Users\1700\Desktop\信息可视化\源数据\国家统计局分省年度数据\国民经济核算\居民消费水平.csv'
    obj1 = pd.read_csv(tmp)
    obj2 = obj1[~obj1['年份'].isin([2018,2019])]#选取含有指定值的行从表中删除：~
    obj3 = obj2.reset_index(drop=True)
    
    tmp1 = 0
    tmp2 = []
    for i in range(0,31):
        tmp2.append(obj2['省份'].tolist()[tmp1])
        tmp1 += 18
    
    z = np.array(obj3['居民消费水平(元)'])
    z_1 = []
    for i in range(0, len(z), 18):#18
        z_1.append(z[i : i+18])#18
    z_1 = np.array(z_1).T
    z_2 = list(z_1)

    tl = Timeline()
    count = 0
    for i in range(2017, 1999, -1):
        bar = (
            Bar(init_opts=opts.InitOpts()) 
            .add_xaxis(tmp2)
            .add_yaxis("分省居民消费水平值数", [int(i) for i in list(z_2[count])], label_opts=opts.LabelOpts(position="right"))
            .set_series_opts(label_opts = opts.LabelOpts(is_show = True,position = 'right'))
            .reversal_axis()
            .set_global_opts(
                            toolbox_opts=opts.ToolboxOpts(),
                            title_opts = opts.TitleOpts("{}年分省居民消费水平值数可视化".format(i),pos_left = '0%'),
                            legend_opts = opts.LegendOpts(pos_right = '10%'),
                            yaxis_opts = opts.AxisOpts(
                                axistick_opts=opts.AxisTickOpts(is_inside=True),
                                axislabel_opts=opts.LabelOpts(font_size=9)#!
                            ),
            )
                            )
        count += 1
        tl.add(bar, "{}年".format(i))
    tl.add_schema(is_auto_play=True,is_rewind_play=True,play_interval = 900)
    return tl
#ff().render("居民消费水平条形动图.html")
```
